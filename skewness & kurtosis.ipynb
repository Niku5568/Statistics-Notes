````python
# Square root transormation on left skewed data in Python:
df['B_Sqrt']=np.sqrt(max(df.iloc[:, 2]+1) - df.iloc[:, 2])
````

````python
#log transform
df['C_log']=np.log(df.iloc[:,0])
````

````python
# Log transformation of negatively (left) skewed data in Python
df['D_log']=np.log(max(df.iloc[:, 2] + 1) -df.iloc[:, 2])
````

In Python, you can handle kurtosis using various techniques, including data transformations and statistical tests. Here's how you can handle kurtosis in Python:

### Data Transformations:

1. **Log Transformation:**
   ```python
   import numpy as np
   import pandas as pd

   # Assuming 'data' is your DataFrame and 'column' is the column with kurtosis
   data['log_transformed'] = np.log(data['column'])
   ```

2. **Square Root Transformation:**
   ```python
   data['sqrt_transformed'] = np.sqrt(data['column'])
   ```
### Data Trimming:

1. **Trimming Extreme Values:**
   - Remove outliers from your data to reduce the impact of heavy tails. You can define a threshold based on z-scores or percentiles.
   ```python
   z_scores = (data['column'] - data['column'].mean()) / data['column'].std()
   data_trimmed = data[(z_scores < 3) & (z_scores > -3)]  # Trimming values within 3 standard deviations
